```python
import matplotlib.pyplot as plt

# Define the accuracy values
accuracy = [0.12, 0.16, 0.23, 0.25, 0.30, 0.32, 0.33, 0.34, 0.36, 0.37, 0.40, 0.42, 0.44, 0.46, 0.50, 0.56, 0.57, 0.55, 0.58, .60, .61]

# Define the x-axis values as integers from 0 to 9
x = range(8000)

# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the line graph
plt.plot(x, accuracy, 'b-', linewidth=2)

# Set the y-axis limits
plt.ylim([0, 1])

# Set the maximum value at 0.82
plt.axhline(y=0.82, color='r', linestyle='--')

# Set the title and axis labels
plt.title('Accuracy of Vectorizer')
plt.xlabel('Number of Runs')
plt.ylabel('Accuracy')

# Show the plot
plt.show()
```
